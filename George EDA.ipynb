```python
def genre_counter(genre):
    chars = [char for char in genre]
    counter = 0
    for char in chars:
        if char == ',':
            counter += 1
        
    return counter
```

```python
genre_counter(anime_info['Genders'][200])
```

```python
anime_info['genre_count'] = anime_info['Genders'].map(genre_counter)
anime_info
```

```python
fig, ax = plt.subplots()

plt.hist(anime_info['genre_count'])
plt.title('Historgram for Number of Genres')
plt.xlabel('Number of Genres')
plt.ylabel('Number of Anime');
```
